# Ejecrcicio para importar datos de una url en csv, y guardar como un dataframe local.
Se obtiene el dato del sitio  ejemplo http://winterolympicsmedals.com/medals.csv





```python
import os

medals_url = "http://winterolympicsmedals.com/medals.csv"
medals_data = pd.read_csv(medals_url)

#Primero importamos la libreria y hacemos conexion con la web de datos
import urllib3
http = urllib3.PoolManager()
r = http.request('GET',medals_url)
print("el estado de la respuesta es %d" %(r.status))
response = r.data

# El objeto response contiene un string binario, se convierte a string UTF-8
str_data = response.decode('utf-8')

# Dividimos el string en array de filas, separando por intros
lines = str_data.split("\n")

#Extrae cabecera de la primera fila
col_names = lines[0].split(",")
n_cols = len(col_names)


# Dicionario vacío para volcar la informacion procesada desde la url externa
counter = 0
main_dict = {}
for col in col_names:
    main_dict[col] = []
    
    
# procesamos fila a fila la informacion para ir llenando el diccionario 
for line in lines:
    #Nos saltamos la primera linea que es la cabecera que ya procesamos
    if(counter>0):
        #dividimos cada string por las comas como elemento separador
       values = line.strip().split(",")
       #añadimos cada valor a la respectiva columna del diccionario
       for i in range(len(col_names)):
           main_dict[col_names[i]].append(values[i])
    counter +=1

print("El data set tiene %d filas y %d columnas"%(counter, n_cols))
       
#concertimos el diccionario procesdo a DATAFRAME y comprobamos que los datos son correctos
medals_df = pd.DataFrame(main_dict)
print(medals_df.head())

#elegimos donde guardarlo (En la carpeta athletes es donde tiene mas sentido por el contexto del analisis)
mainpath = "C:/orlando/repo_github/python-ml-course/datasets/"

filename = "athletes/orly_downloaded_medals."
fullpath = os.path.join(mainpath,filename) 


#Guardamos en csv, json, excel
medals_df.to_csv(fullpath + "csv")
medals_df.to_json(fullpath + "json")
medals_df.to_excel(fullpath + "xls")
       
       
       


```

```python

def downloadFromUrl(url,filename, sep=",",delim = "\n", encoding="utf-8",
                   mainpath = "C:/orlando/repo_github/python-ml-course/datasets/"):
    #filename = "athletes/orly_downloaded_medals"
    import os

    medals_url = "http://winterolympicsmedals.com/medals.csv"
    medals_data = pd.read_csv(medals_url)

    #Primero importamos la libreria y hacemos conexion con la web de datos
    import urllib3
    http = urllib3.PoolManager()
    r = http.request('GET',url)
    print("el estado de la respuesta es %d" %(r.status))
    response = r.data

    # El objeto response contiene un string binario, se convierte a string UTF-8
    str_data = response.decode(encoding)

    # Dividimos el string en array de filas, separando por intros
    lines = str_data.split(delim)

    #Extrae cabecera de la primera fila
    col_names = lines[0].split(sep)
    n_cols = len(col_names)


    # Dicionario vacío para volcar la informacion procesada desde la url externa
    counter = 0
    main_dict = {}
    for col in col_names:
        main_dict[col] = []


    # procesamos fila a fila la informacion para ir llenando el diccionario 
    for line in lines:
        #Nos saltamos la primera linea que es la cabecera que ya procesamos
        if(counter>0):
            #dividimos cada string por las comas como elemento separador
           values = line.strip().split(sep)
           #añadimos cada valor a la respectiva columna del diccionario
           for i in range(len(col_names)):
               main_dict[col_names[i]].append(values[i])
        counter +=1

    print("El data set tiene %d filas y %d columnas"%(counter, n_cols))

    #concertimos el diccionario procesdo a DATAFRAME y comprobamos que los datos son correctos
    df = pd.DataFrame(main_dict)
    print(medals_df.head())

    #elegimos donde guardarlo (En la carpeta athletes es donde tiene mas sentido por el contexto del analisis)
    #mainpath = "C:/orlando/repo_github/python-ml-course/datasets/"

    #filename = "athletes/orly_downloaded_medals"
    fullpath = os.path.join(mainpath,filename) 


    #Guardamos en csv, json, excel
    df.to_csv(fullpath + ".csv")
    df.to_json(fullpath + ".json")
    df.to_excel(fullpath + ".xls")
    
    print("Archivo guardado en: " + fullpath)
    
    return df
```
