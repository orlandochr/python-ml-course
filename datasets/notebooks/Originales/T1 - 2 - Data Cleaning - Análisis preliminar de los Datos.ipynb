```python
### Crea variables dummies a partir de la columna sex, divide en fem y masculino,
#### borra la columna sexo y concatena las columnas sex_female sex_male

def createDummies(df, var_name):
    dummy = pd.get_dummies(df[var_name], prefix=var_name)
    df = df.drop(var_name, axis = 1)
    df = pd.concat([df, dummy], axis = 1)
    return df
```
